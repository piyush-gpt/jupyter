```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf


df=pd.read_csv("C:\\Users\\lenovo\\Downloads\\Churn_Modelling.csv")
df.head()
```

```python
# seperating dependent and independent features
X=df.iloc[:,3:13]
Y=df.iloc[:,13]
X.head()
```

```python
# hadling cloumns having classification (Geography and gender)
geography=pd.get_dummies(X["Geography"],drop_first=True,dtype=int)# by drop_first, we are removing first column as both spain and germany are zero means it is france
gender=pd.get_dummies(X["Gender"],drop_first=True,dtype=int)
geography.head()
```

```python
# now concating the new columns

X=X.drop(["Geography","Gender"], axis=1)# droping old columns
```

```python
X=pd.concat([X,geography,gender], axis=1)
X.head()

```

```python
# splitting traning and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# creating ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, ELU, ReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()
```

```python
# adding input layer

classifier.add(Dense(7, input_shape=(11,), activation="relu"))# created first hidden layer having 7 neurons and input layer is automatically created
```

```python
#adding more hidden layer 
classifier.add(Dense(6, activation="relu"))
```

```python
#adding output layer
classifier.add(Dense(1, activation="sigmoid"))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#to stop early when their is no improvement in accuracy
early_stopping=tf.keras.callbacks. EarlyStopping(
monitor="val_loss",
min_delta=0.0001,
patience=20,
verbose=1,
mode="auto",
baseline=None,
restore_best_weights=False,
)    
```

```python
model_history=classifier.fit(X_train, Y_train, validation_split=0.33, batch_size=10, epochs=50,callbacks=early_stopping)
```

```python
X_predict=classifier.predict(X_test)
```
